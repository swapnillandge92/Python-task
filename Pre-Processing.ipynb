```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.preprocessing as LableEncoder
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
df_train = pd.read_csv("D:\\Data Science\\netzwrek\\Data\\train.csv")
df_test = pd.read_csv("D:\\Data Science\\netzwrek\\Data\\test.csv")
```

```python
df_train.dropna(axis = 0, inplace = True)
```

```python
label_encoder = preprocessing.LabelEncoder()

df_train['data3']= label_encoder.fit_transform(df_train['data3'])
df_train['data9']= label_encoder.fit_transform(df_train['data9'])
df_train['data10']= label_encoder.fit_transform(df_train['data10'])

df_test['data3']= label_encoder.fit_transform(df_test['data3'])
df_test['data9']= label_encoder.fit_transform(df_test['data9'])
df_test['data10']= label_encoder.fit_transform(df_test['data10'])


df_train['data1']= label_encoder.fit_transform(df_train['data1'])
df_train['data2']= label_encoder.fit_transform(df_train['data2'])
df_train['data4']= label_encoder.fit_transform(df_train['data4'])

df_test['data1']= label_encoder.fit_transform(df_test['data1'])
df_test['data2']= label_encoder.fit_transform(df_test['data2'])
df_test['data4']= label_encoder.fit_transform(df_test['data4'])

df_train['data5']= label_encoder.fit_transform(df_train['data5'])
df_train['data6']= label_encoder.fit_transform(df_train['data6'])
df_train['data7']= label_encoder.fit_transform(df_train['data7'])
df_train['data8']= label_encoder.fit_transform(df_train['data8'])

df_test['data5']= label_encoder.fit_transform(df_test['data5'])
df_test['data6']= label_encoder.fit_transform(df_test['data6'])
df_test['data7']= label_encoder.fit_transform(df_test['data7'])
df_test['data8']= label_encoder.fit_transform(df_test['data8'])
```

```python
X = df_train.drop(['output'], axis = 1, inplace = False )
y = df_train['output']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, random_state = 10)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
df_predict = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
df_predict
```

```python
r2_score(y_test, y_pred)
```
