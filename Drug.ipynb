```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing 

from sklearn.model_selection import train_test_split

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train=pd.read_csv("D:\\Data Science\\netzwrek\\Assignement\\drug_consumption.csv",encoding='latin1')
```

```python
train.dropna(axis = 0, inplace = True)
```

```python
df_train = train
```

```python
label_encoder = preprocessing.LabelEncoder() 
  

df_train['Alcohol']= label_encoder.fit_transform(df_train['Alcohol']) 
df_train['Amphyt']= label_encoder.fit_transform(df_train['Amphyt'])
df_train['Amyl']= label_encoder.fit_transform(df_train['Amyl']) 
   
df_train['Benzos']= label_encoder.fit_transform(df_train['Benzos']) 
df_train['Caff']= label_encoder.fit_transform(df_train['Caff'])
df_train['cannabis']= label_encoder.fit_transform(df_train['cannabis'])  


df_train['choc']= label_encoder.fit_transform(df_train['choc']) 
df_train['coke']= label_encoder.fit_transform(df_train['coke'])
df_train['Crack']= label_encoder.fit_transform(df_train['Crack']) 
   
df_train['Esctasy']= label_encoder.fit_transform(df_train['Esctasy']) 
df_train['Heroin']= label_encoder.fit_transform(df_train['Heroin'])
df_train['ketamine']= label_encoder.fit_transform(df_train['ketamine'])  

df_train['Legalh']= label_encoder.fit_transform(df_train['Legalh']) 
df_train['LSD']= label_encoder.fit_transform(df_train['LSD'])
df_train['Meth']= label_encoder.fit_transform(df_train['Meth']) 
   
df_train['mushrooms']= label_encoder.fit_transform(df_train['mushrooms']) 
df_train['Nicotine']= label_encoder.fit_transform(df_train['Nicotine'])
df_train['Semer']= label_encoder.fit_transform(df_train['Semer']) 

df_train['VSA']= label_encoder.fit_transform(df_train['VSA'])
```
